# fastMRI dataset statistics

The created `csv` files contain meta information. We extract this information here to get statistics about the datasets, e.g. acquisition type (`type`), MR scanner (`systemModel`), number of Phase Encoding (PE) lines (`enc_y`), acceleration factor (`acc`).

```python
import pandas as pd
import os

fastmri_root = os.getenv('FASTMRI')
dataset = 'multicoil_train'

df = pd.read_csv(f'{fastmri_root}/datasets/{dataset}.csv')
df.groupby(['type', 'systemVendor', 'systemModel', 'systemFieldStrength_T','enc_y']).count()
```
